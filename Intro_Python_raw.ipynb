## Lists: Your Go-To for Storing Collections

Lists are used to store multiple items in a single variable. They are created using square brackets `[]`.

```python
inventory = ["sword", "shield", "potion"]
print(inventory)
print(type(inventory))
```

### List Length

You can get the number of items in a list with the `len()` function.

```python
print(len(inventory))
```

### Accessing Items

You can access items in a list by their index.

```python
print(inventory[1])
```

### Negative Indexing

You can also use negative indexing to access items from the end of the list.

```python
print(inventory[-1]) #last item on the list
```

### Changing Items

You can change the value of an item by referring to its index.

```python
inventory[1] = "magic shield"
print(inventory)
```

### Adding Items

You can add an item to the end of the list with the `append()` method.

```python
inventory.append("gold coin")
print(inventory)
```

You can also insert an item at a specific position with the `insert()` method.

```python
inventory.insert(1, "healing herb")
print(inventory)
```

### Removing Items

You can remove an item with the `remove()` method.

```python
inventory.remove("sword")
print(inventory)
```

You can also remove an item at a specific index with the `pop()` method.

```python
inventory.pop(1)
print(inventory)
```

### Looping Through a List

You can loop through the items in a list with a `for` loop.

```python
for item in inventory:
  print(item)
```

## Conditional Logic: If ... Else

Python supports the usual logical conditions from mathematics:

*   Equals: `a == b`
*   Not Equals: `a != b`
*   Less than: `a < b`
*   Less than or equal to: `a <= b`
*   Greater than: `a > b`
*   Greater than or equal to: `a >= b`

You can use these conditions to control the flow of your program.

### The `if` statement

```python
a = 33
b = 200
if b > a:
  print("b is grater than a")
```

### The `elif` statement

The `elif` keyword is Python's way of saying "if the previous conditions were not true, then try this condition".

```python
a = 33
b = 33

if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

### The `else` statement

The `else` keyword catches anything which isn't caught by the preceding conditions.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

## Loops: Repeating Yourself (the Smart Way)

Loops are used to execute a block of code multiple times.

### The `while` loop

The `while` loop executes a set of statements as long as a condition is true.

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

You can use the `break` statement to stop the loop even if the condition is still true.

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

### The `for` loop

The `for` loop is used for iterating over a sequence (like a list, tuple, dictionary, set, or string).

```python
fruits = ["apple", "banana", "cherry"]
for fruits in fruits:
  print(fruits)
```

## Functions: Reusable Blocks of Code

A function is a block of code that only runs when it is called. You can pass data, known as parameters, into a function, and it can return data as a result.

### Creating a Function

You define a function using the `def` keyword.

```python
def my_funtion():
  print("Hello from an function")
```

### Calling a Function

To call a function, you use its name followed by parentheses.

```python
my_funtion()
```

### Arguments

You can pass information into functions as arguments.

```python
def greet(name):
  print("Hello, " + name + "!")

greet("Aragorn")
greet("Legolas")
greet("Gimli")
```

### Return Values

To let a function return a value, use the `return` statement.

```python
def my_function(x):
  y = 5 * x
  return y

print(my_function(3))
print(my_function(5))
print(my_function(9))
```

## Exercise: Your First Python Game!

Let's create a simple "Guess the Number" game. The computer will think of a number between 1 and 100, and you'll have to guess it!

**Your Mission:**
1.  **Generate a Random Number:** Use the `random` module to generate a random number between 1 and 100.
2.  **Get User Input:** Ask the user to guess the number.
3.  **Use a Loop:** Keep asking the user for a guess until they get it right.
4.  **Give Hints:** Tell the user if their guess is too high or too low.
5.  **Congratulate the Winner:** Once the user guesses the number, congratulate them!

```python
import random

secret_number = random.randint(1,100)
print("Guess the number")
print("I'm thinking of a number between 1 and 100")
while True:
  guess = input("Enter your guess: ")
  if not guess.isdigit():
    print("Please enter a valid number.")
    continue

  guess = int(guess)

  if guess < secret_number:
    print("guess is too low")

  elif guess > secret_number:
    print("guess is too high")

  else:
    print("Congratulations! you got it")
    break

#Hints
#While the users guest is not the secret number -> another input
#if the guess is too low -> give an hint
#or if the guess is too high -> give an hint
```

```python
import random

secret_number = random.randint(1,100)
guess = 0

while guess != secret_number:
  guess = int(input("Guess a number between 1 and 100:"))

  if guess < secret_number:
    print("Too Low!, Try again.")
  elif guess > secret_number:
    print("Too high! Try again")

  print(f"Congrats! you guessed the number: {secret_number}")
```
