### 1. Generate Synthetic Data
Using NumPy, create an array `X` with 100 random numbers between 0 and 10. Create a corresponding array `y` such that:


**$y = 2X + 3 + \text{random noise}$**


- Use `np.random.rand` to generate the random values for `X` and `y`.
- The random noise should be small (e.g., normally distributed noise).

```python
# Import necessary libraries
import numpy as np

np.random.seed(42)  # For reproducibility

# 100 random numbers between 0 and 10
X = np.random.rand(100) * 10

# Random noise
noise = np.random.randn(100)

# Linear relation with some noise
y = 2 * X + 3 + noise
```

### 2. Visualize the Data
Use Matplotlib to create a scatter plot of `X` and `y`. The goal here is to visualize the relationship between the input variable `X` and the output `y`.

```python
import matplotlib.pyplot as plt

# Sacatter plot of x and y
plt.scatter(X, y, color='blue', alpha=0.7)

# Label the x axis with x and the y axis with y
plt.xlabel("X")
plt.ylabel("y")
plt.title("Sactter plot of X and y")

# Show the plot
plt.show()
```

### 3. Implement Simple Linear Regression

Write a function `linear_regression(X, y)` that performs simple linear regression by computing the slope `m` and the intercept `b` of the best-fit line using the formulas:

$\displaystyle m = \frac{n \sum(X \cdot y) - \sum(X) \sum(y)}{n \sum(X^2) - (\sum(X))^2}$

$\displaystyle b = \frac{\sum(y) - m \sum(X)}{n}$

Where `n` is the number of data points.

```python
def linear_regression(X, y):
  """
  Perform simple linear regression to find the slope (m) and intercept (b) of the best-fit line.

  Parameters:
  X (array-like): Input features.
  y (array-like): Target values.

  Returns:
  tuple: Slope (m) and intercept (b) of the best-fit line.
  """
  # Calculate the number of data points
  n = len(X)
  # Calculate the slope (m) using the fomula
  m = (n * np.sum(X * y) - np.sum(X) * np.sum(y)) / (n * np.sum(X**2) - (np.sum(X))**2)
  # Calculate the intercept (b) using the formula
  b = (np.sum(y) - m * np.sum(X)) / n
  # return the slope (m) and the intercept (b)
  return m, b
```

```python
 # Calculate the slope and intercept using the linear_regression function
m, b = linear_regression(X, y)
# Print the calculated slope (m) using f-string: Slope (m): m
print(f"Slope (m): {m}")
# Print the calculated intercept (i) using f-string: Intercept (b): b
print(f"Intercept (b): {b}")
```
